```python
# Import Packages
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import Dataset

# Define File Path
file_path = 'C:\\Users\\18655\\Downloads\\auto+mpg\\auto-mpg.txt'
column_names = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name']
df = pd.read_csv(file_path, sep='\s+', names = column_names)

# Check if 'horsepower' column contains a question mark
condition = df['horsepower'] == '?'

# Verify the data type of the condition
print("Data type of the condition:", condition.dtype)

# Use the boolean condition to filter the DataFrame and remove rows with question marks
df_cleaned = df[~condition]

# Change to Float
print(df_cleaned.head(33))
```

```python
# Data Pre-Processing

# Identify NaNs
nan_mask = df_cleaned.isna()

# Count NaNs
nan_count = nan_mask.sum()

# Print Count
print(nan_count)

# No Missing Data- Perform Analysis
```

```python
df_cleaned['horsepower'] = df_cleaned['horsepower'].astype(float)
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
scaled_displacement = scaler.fit_transform(df_cleaned[['displacement']])
scaled_horsepower = scaler.fit_transform(df_cleaned[['horsepower']])
scaled_weight = scaler.fit_transform(df_cleaned[['weight']])
scaled_acceleration = scaler.fit_transform(df_cleaned[['acceleration']])
scaled_features = np.column_stack((scaled_displacement, scaled_horsepower, scaled_weight, scaled_acceleration))
scaled_columns = ['displacement','horsepower','weight','acceleration']

# Convert the scaled feature back to a DataFrame
scaled_df = pd.DataFrame(scaled_features, columns=scaled_columns)
```

```python
# Drop Columns
columns_to_drop = ['displacement','horsepower','weight','acceleration']  # List of column names to remove
df_cleaned = df_cleaned.drop(columns=columns_to_drop, axis=1)
df_cleaned = df_cleaned.reset_index(drop=True)

# Add Scaled DF
final_cleaned_scaled_df = pd.concat([df_cleaned, scaled_df], axis=1)

# Print Final DF
print(final_cleaned_scaled_df)
```

```python
# Correlation Coefficient Matrix

correlation_coefficient_acceleration = final_cleaned_scaled_df['acceleration'].corr(final_cleaned_scaled_df['mpg'])
correlation_coefficient_cylinders = final_cleaned_scaled_df['cylinders'].corr(final_cleaned_scaled_df['mpg'])
correlation_coefficient_displacement = final_cleaned_scaled_df['displacement'].corr(final_cleaned_scaled_df['mpg'])
correlation_coefficient_horsepower = final_cleaned_scaled_df['horsepower'].corr(final_cleaned_scaled_df['mpg'])
correlation_coefficient_model_year = final_cleaned_scaled_df['model_year'].corr(final_cleaned_scaled_df['mpg'])
correlation_coefficient_origin = final_cleaned_scaled_df['origin'].corr(final_cleaned_scaled_df['mpg'])
correlation_coefficient_weight = final_cleaned_scaled_df['weight'].corr(final_cleaned_scaled_df['mpg'])

print("Acceleration Correlation is: ", correlation_coefficient_acceleration)
print("Cylinders Correlation is: ", correlation_coefficient_cylinders)
print("Displacement Correlation is: ", correlation_coefficient_displacement)
print("Horsepower Correlation is: ", correlation_coefficient_horsepower)
print("Model Year Correlation is: ", correlation_coefficient_model_year)
print("Origin Correlation is: ", correlation_coefficient_origin)
print("Weight Correlation is: ", correlation_coefficient_weight)
```

```python
# Create a dictionary with the correlations
correlations = {
    'Acceleration': 0.4233285369027872,
    'Cylinders': -0.7776175081260214,
    'Displacement': -0.8051269467104581,
    'Horsepower': -0.7784267838977762,
    'Model Year': 0.5805409660907855,
    'Origin': 0.5652087567164612,
    'Weight': -0.8322442148315752
}

# Convert the correlations dictionary to a DataFrame
corr_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Correlation'])

# Create a pivot table for the heatmap
corr_matrix = corr_df.pivot_table(index='Feature', columns='Feature', values='Correlation')

# Set up the heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  # Adjust font size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Train Test Split

# Define X
X = final_cleaned_scaled_df[['cylinders','displacement','horsepower','weight']]

# Define Y (Target Variable)
Y = final_cleaned_scaled_df['mpg']

# Split Train/Test
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size=.2,random_state=10)
```

```python
# Create Linear Regression Model

# Initialize Model
model = LinearRegression()
model.fit(X_Train, Y_Train)
y_predictions = model.predict(X_Test)

# Find MSE
mse = mean_squared_error(Y_Test, y_predictions)
r2 = r2_score(Y_Test, y_predictions)

# Print 
print("Linear MSE: ", mse)
print("R-Squared: ", r2)
```

```python
# Train Random Forest Model

# Create Loop for Optimizing n_estimators

results = []
for n_estimators in range(1, 101):
    model = RandomForestRegressor(n_estimators=n_estimators, random_state=42)
    model.fit(X_Train, Y_Train)
    y_pred = model.predict(X_Test)
    mse = mean_squared_error(Y_Test, y_pred)
    results.append((n_estimators, mse))

# Find the n_estimators with the lowest MSE
best_n_estimators, best_mse = min(results, key=lambda x: x[1])

print(f"Best n_estimators: {best_n_estimators}")
print(f"Corresponding MSE: {best_mse}")

# Optimal Number of n_estimators is 19 with MSE 12.34, this is lower than linear regression
```

```python
# Train Support Vector Machine Model
 
# Initialize and train SVR model
model = SVR(kernel='rbf')
model.fit(X_Train, Y_Train)
y_pred_SVR = model.predict(X_Test)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y_Test, y_pred_SVR)
print(f"Mean Squared Error: {mse}")

# MSE of 16.05 is higher than decision trees but lower than linear regression
```
